```python
# dropping non important columns that we wont be needing
# renaming Country/region to country
dfnew = df.rename(columns={'Country/Region' : 'Country'})
dfup = dfnew.drop(dfnew.columns[[0,2,3]], axis=1)
dfup.head()
```

```python
#grouping countries
df_groupbycountry = dfup.groupby("Country").sum()
df_groupbycountry.head(10)
```

displaying the top 5 countries as of 11/28/21 to be able to identify the top 3

```python
df_sorted = df_groupbycountry.sort_values('1/30/22', ascending=False)
df_sorted.head(10)
# from the output the top three countries can be seen to be US,India and Brazil
```

new dataframe with only the top three infected countries. dropping the rest

```python
#Selecting top 3 countries
df_top3 = df_sorted.query("Country == 'US' | Country == 'India' | Country == 'Brazil'")
df_top3
```

visualization for the top three countries

```python
import matplotlib.pyplot as plt
#Figure size
plt.figure(figsize=(9, 4))

#Plot
df_top3['1/30/22'].plot.bar(align='center',
                   alpha=0.5, color=['green', 'orange', 'blue'], ecolor='black', capsize=10)

#Labels
plt.ylabel('Total COVID cases\n as of 30th Janurary 22\n', fontsize=12);
plt.xlabel("")
plt.title('\nChart 1: Top 3 COVID infected countries in the world\n', fontsize=15)

plt.show()
```

Representing the dates in the form of a week number e.g week 1 for 22 January 2020 to 30th January 2022

number of weeks from january 22 2020 to november 28 2021 are: 96 weeks and 4 days

going with 96 weeks

```python
##Creating a new dataframe with only data until 24 November 2021 (dropping last 4 columns)
df_top3_96weeks = df_top3.drop(columns=['11/25/21', '11/26/21', '11/27/21', '11/28/21'], axis=1)
df_top3_96weeks

##Grouping values in 7 days and taking the last value (total of the week)
df_top3_cumulative = df_top3_96weeks.groupby([[i//7 for i in range(0,673)]], axis = 1).last()

##Renaming the title of each column starting from 1 (1, 2, 3... instead of 0, 1, 2...)
df_top3_cumulative.columns = [(i+1) for i,_ in enumerate(df_top3_cumulative.columns)]

##Total cases per week of Brazil, India and the US
df_top3_cumulative
```

visualizing how cases have evolved from week1 to week 96 in the top 3 infected countries

```python
import seaborn as sns
import numpy as np
##Reseting index to include Country names in chart
df2 = df_top3_cumulative.reset_index()

#Preparing values
df3 = df2.melt(id_vars='Country', var_name='Week', value_name='Cumulative cases')

##size of chart
plt.figure(figsize=(15, 6))

#Load lineplot with data
df4 = sns.lineplot(x='Week', y='Cumulative cases', hue='Country', data=df3)

df4.set_title("\nChart 2: COVID Weekly cumulative cases in US, India and Brazil\n 22 January 2020 - 24 November 2021 \n", fontsize=15)
plt.xlabel("\nWeek number")
plt.xticks(np.arange(97), rotation=90)

plt.show()

# analysis
# from the lineplot we can see that US has had an increase in covid 19 cases
# india and brazil covid cases have been contant and might be descreasing
# this migth change with the new covid variant omicron
```

calculating new cases per week inorder to perform a better predictive model, clustering and graph analytics

```python
#Difference between each column and previous column result in the new cases for each week
df_top3_newcases = df_top3_cumulative[df_top3_cumulative.columns[0:96]].diff(axis=1)

#Fixing first week values
df_top3_newcases.at[['Brazil','India'],1]=0
df_top3_newcases.at[['US'],1]=5

df_top3_newcases
```

Visualisations for weekly NEW CASES from Week 1 to Week 66 for US, India and Brazil

```python
##Reseting index to include Country names in chart
df5 = df_top3_newcases.reset_index()

##Melting
df6 = df5.melt(id_vars='Country', var_name='Week', value_name='Weekly new cases')

##size of chart
plt.figure(figsize=(15, 6))

##Creating chart
df7 = sns.lineplot(x='Week', y='Weekly new cases', hue='Country', data=df6)

##Title
df7.set_title("\nChart 3: COVID Weekly new cases in US, India and Brazil from\n 22 January 2020 - 24th November 2021\n", fontsize=15)

plt.xlabel("\nWeek number")
plt.xticks(np.arange(97), rotation=90)

plt.show()

# analysis
# from the lineplot we can see that US cases have been increasing
# whereas for India and Brazil cases have been dropping and they should be on their way to recovery
```

Predictive Modelling

Linear Regression

```python
# importing scikit learn libraries to use
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# converting columns into rows using the transpose function
# swapping columns and rows of cumulative cases data called "df_top3_cumulative"
df_newcases = df_top3_newcases.transpose()

#Resetting index
df_newcases = df_newcases.reset_index()

#assigning new names of columns
df_newcases.columns = ['Week number', 'US', 'India', 'Brazil']

df_newcases
```

linear regression model for each of the top three infected countries

Linear Regression for US

Scatterplot for US new covid cases

```python
plt.figure(figsize=(15, 6))
plt.scatter(df_newcases['Week number'], df_newcases['US'])
plt.xticks(rotation=90)
plt.title("\nChart 8: Scatter Plot of COVID Weekly new cases in US\n22 January 2020 - 24 November 2021 \n", fontsize=15)
plt.ylabel("New cases per week")
plt.xlabel("\nWeek number")
plt.xticks(np.arange(97), rotation=90)
plt.show()
```

Training and Testing Linear Regression model for Us

```python
#Assigning X and Y to independent and dependent variables, respectively
X = df_newcases[['Week number']] #Independent (Week number)
y = df_newcases['US'] #Dependent (Count of infection per week)

#Selecting train and test data randomly. 80% train, 20% test
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size =0.2)

#Creating Linear Regression object 'regr'
regr = LinearRegression()

#Training the model
regr.fit(X_train,y_train)

# Make predictions using the testing set X_test
y_pred = regr.predict(X_test)

# Coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))

# Plot outputs
plt.figure(figsize=(15, 6))

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='green', linewidth=2)

#Plot title
plt.title("\nChart 8: Linear Regression - Predictive Modelling\nCOVID Weekly new cases in US\n22 January 2020 - 24 November 2021 \n", fontsize=15)
plt.ylabel("New cases per week")
plt.xlabel("\nWeek number")
plt.xticks(np.arange(97), rotation=90)

plt.show()

# Analysis 
# After multiple iterations, the COEFFICIENT OF DETERMINATION is approx. -0.7, the variance between the model and the actual data is very high.
# the line shows that the cases have been increasing
```

# Going with US because it has high variance
#CLUSTERING

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# new dataframe
df_us= df_newcases.drop(df_newcases[['India', 'Brazil']], axis=1)
df_us
```

 Using Elbow method to determine the best K value (number of clusters)

```python
#Range
k_rng = range(1,11)

#Empty list SSE
sse = []

#Inertia_ function
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_us[['Week number', 'US']])
    sse.append(km.inertia_)
```

```python
#Plotting Elbow Method
plt.figure(figsize=(12, 6))
plt.plot(range(1, 11), sse)
plt.title('Chart 9: Elbow Method for determinying number of clusters K\n US Clustering model\n', fontsize=15)
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared error')
plt.show()
```

```python
# the best k value is 3
#Assigning number of clusters
km = KMeans(n_clusters=3)
km
```

```python
#Creating cluster predictions
y_predicted = km.fit_predict(df_us[['Week number', 'US']])
y_predicted
```

```python
#Creating cluster column in dataframe
df_us['cluster'] = y_predicted
df_us
```

```python
#Plotting scatters for the 3 clusters:
df1 = df_us[df_us.cluster==0]
df2 = df_us[df_us.cluster==1]
df3 = df_us[df_us.cluster==2]

plt.figure(figsize=(12, 6))
plt.title('Chart 10: K-Means Clustering for US COVID weekly new cases \n22 January 2020 - 24 November 2021\nFirst iteration\n', fontsize=15)

plt.scatter(df1['Week number'], df1['US'], color = 'green')
plt.scatter(df2['Week number'], df2['US'], color = 'red')
plt.scatter(df3['Week number'], df3['US'], color = 'black')

plt.xlabel('Week number')
plt.ylabel('Number of cases')

plt.show()
```

As shown in the scatter plot, these 3 clusters are not too accurate as there is green areas between the three black areas. need to improve and stabilise the clusters.


```python
#Scale 0 to 1 for Week number and US columns, using MinMaxScaler() function:
scaler = MinMaxScaler()

scaler.fit(df_us[['US']])
df_us[['US']] = scaler.transform(df_us[['US']])

scaler.fit(df_us[['Week number']])
df_us[['Week number']] = scaler.transform(df_us[['Week number']])

df_us
```

```python
#Creating new cluster predictions with scaled values (second iteration)
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df_us[['Week number', 'US']])
y_predicted
```

```python
#Displaying new cluster predictions (improved)
df_us['cluster'] = y_predicted
df_us
```

```python
#Cluster centers
km.cluster_centers_
```

Plotting new K Means Clustering 

```python
df1 = df_us[df_us.cluster==0]
df2 = df_us[df_us.cluster==1]
df3 = df_us[df_us.cluster==2]

plt.figure(figsize=(12, 6))
plt.title('Chart 11: K-Means Clustering for US COVID weekly new cases \n22 January 2020 - 24 November 2021\nSecond (improved) iteration\n', fontsize=15)

plt.scatter(df1['Week number'], df1['US'], color = 'green')
plt.scatter(df2['Week number'], df2['US'], color = 'blue')
plt.scatter(df3['Week number'], df3['US'], color = 'orange')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'purple', marker='*', s=250)

plt.xlabel('Week number (scaled)')
plt.ylabel('Number of cases (scaled)')

plt.show()
```
